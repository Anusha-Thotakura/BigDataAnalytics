```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("SavingsAnalysis").getOrCreate()
```

```python
df = spark.read.csv("C:\\Users\\Anusha\\Documents\\data.csv", header=True, inferSchema=True)
```

```python
df.printSchema()
```

 Total records

```python
print("Total rows:", df.count())
```

1. Income Distribution

```python
import matplotlib.pyplot as plt

income_data = df.select("Income").toPandas()

plt.figure(figsize=(8,5))
plt.hist(income_data["Income"], bins=30, color="skyblue", edgecolor="black")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

2. Average Income & Savings

```python

from pyspark.sql.functions import avg

df.select(avg("Income").alias("AvgIncome"),
          avg("Desired_Savings").alias("AvgDesiredSavings")).show()
```

3. Income vs Desired Savings

```python
income_savings = df.select("Income", "Desired_Savings").toPandas()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.scatterplot(x="Income", y="Desired_Savings", data=income_savings, alpha=0.5)
plt.title("Income vs Desired Savings")
plt.show()
```

4. Average Savings by Occupation

```python
occupation_income = df.groupBy("Occupation").avg("Income").toPandas()
plt.figure(figsize=(10,5))
plt.bar(occupation_income["Occupation"], occupation_income["avg(Income)"], color="purple")
plt.title("Average Income by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Average Income")
plt.xticks(rotation=45)
plt.show()
```

5.Expense Distribution

```python
expense_cols = ["Groceries","Transport","Eating_Out","Entertainment","Utilities",
                "Healthcare","Education","Miscellaneous"]

expenses_pd = df.select(expense_cols).toPandas()

plt.figure(figsize=(10,6))
sns.boxplot(data=expenses_pd)
plt.xticks(rotation=45)
plt.title("Distribution of Expenses by Category")
plt.show()
```

6.Savings Percentage Distribution

```python
savings_pct = df.select("Desired_Savings_Percentage").toPandas()

plt.figure(figsize=(8,5))
sns.histplot(savings_pct["Desired_Savings_Percentage"], bins=30, kde=True)
plt.title("Desired Savings Percentage Distribution")
plt.xlabel("Savings %")
plt.ylabel("Frequency")
plt.show()
```

7.Correlation Heatmap

```python
numeric_cols = ["Income","Age","Desired_Savings","Disposable_Income"] + expense_cols
corr_pd = df.select(numeric_cols).toPandas().corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_pd, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

8.Age Distributionstribution

```python
age_data = df.select("Age").toPandas()
plt.figure(figsize=(8,5))
plt.hist(age_data["Age"], bins=20, color="teal", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

9.Average Income by Education

```python
edu_income = df.groupBy("Education").avg("Income").toPandas()

plt.figure(figsize=(10,5))
plt.bar(edu_income["Education"], edu_income["avg(Income)"], color="crimson")
plt.title("Average Income by Education")
plt.xlabel("Education Level")
plt.ylabel("Average Income")
plt.xticks(rotation=45)
plt.show()
```

10.Average Income by Occupation

```python
occupation_income = df.groupBy("Occupation").avg("Income").toPandas()

plt.figure(figsize=(10,5))
plt.bar(occupation_income["Occupation"], occupation_income["avg(Income)"], color="purple")
plt.title("Average Income by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Average Income")
plt.xticks(rotation=45)
plt.show()
```

11.Education vs Average Income

```python
education_income = df.groupBy("Education").avg("Income").toPandas()
plt.figure(figsize=(10,5))
plt.bar(education_income["Education"], education_income["avg(Income)"], color="blue")
plt.title("Average Income by Education")
plt.xlabel("Education")
plt.ylabel("Average Income")
plt.xticks(rotation=45)
plt.show()
```

12.Income vs Age (Scatter)

```python
income_age = df.select("Income", "Age").toPandas()
plt.figure(figsize=(8,5))
plt.scatter(income_age["Age"], income_age["Income"], alpha=0.6, color="brown")
plt.title("Income vs Age")
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```
